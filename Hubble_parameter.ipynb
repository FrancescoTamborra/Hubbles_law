```python
# Fitting Hubble's data with a linear regression model to find the 
# Hubble's parameter, K (i.e. the slope of the best fit line).

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#import os
#os.getcwd()
```

```python
df = pd.read_csv("hubble_data.csv") # loading the data
df # showing the data
```

```python
x = df['distance'][:, np.newaxis] # x needs to be 2D for LinearRegression()
y = df['recession_velocity']
```

```python
reg = LinearRegression() 
result = reg.fit(x,y)
slope = "{0:.2f}".format(reg.coef_[0]) # we take the float coefficient (i.e. the slope of the 
# straight line which best fits the data) from the array 'coef_' and we round it at 2 decimal 
# digits converting it into a string.

# print(slope)
```

```python
test_x = x # we want to fit the whole data
y_pred = reg.predict(test_x)
```

```python
fig = plt.figure(figsize=(12,10))
plt.rcParams.update({'font.size' : 15})
plt.title("Hubble's law")
plt.xlabel("distance (Mpc)")
plt.ylabel("recession velocity (Km/s)")
plt.plot(x, y, 'ok', label='data')
plt.plot(x, y_pred, 'r', label='linear fit')
plt.text(1.40, -200, "K = {} (Km/s)/Mpc".format(slope), color='k')
legend = plt.legend(loc=(0.72, 0.15), shadow=True)
```

```python
# The slope of the line which best fits the data (coef_) is 
# the Hubble's parameter, K (nowadays represented as H_0).
# The value found is consistent with the result from the original Hubble's paper:
# https://www.pnas.org/content/15/3/168.full
```
